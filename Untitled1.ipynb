```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import sklearn.preprocessing as skl_pre
import sklearn.linear_model as skl_lm
import sklearn.discriminant_analysis as skl_da
import sklearn.neighbors as skl_nb
import sklearn.model_selection as skl_ms

#from IPython.display import set_matplotlib_formats
#set_matplotlib_formats('png')
from IPython.core.pylabtools import figsize
figsize(10, 6) # Width and hight
#plt.style.use('seaborn-white')
```

```python
url = 'https://raw.githubusercontent.com/Marksmug/Machine-learning/main/train.csv'
data = pd.read_csv(url, na_values='?', dtype={'ID': str}).dropna().reset_index()

data.head()
```

```python
pd.plotting.scatter_matrix(data.iloc[:, 2:14],figsize=(10,10))
plt.show()
```

```python
#split train set and test set randomly
np.random.seed(1)
trainI = np.random.choice(data.shape[0], size = 650, replace=False)
train_index = data.index.isin(trainI)
train = data.iloc[train_index]
test = data.iloc[~train_index]
```

```python
#compute accuracy with one variable
Y_train = train['Lead']
Y_test = test['Lead']

print(Y_train.shape)
for i in range(1, 14):
  X_train = train.iloc[:, [i]]
  X_test = test.iloc[:, [i]]

  model = skl_da.QuadraticDiscriminantAnalysis()
  model.fit(X_train, Y_train)
  prediction = model.predict(X_test)

  accracy = np.mean(prediction == Y_test)
  print(f"Accuracy with {i} column is :{accracy}")
```

```python
#compute accuracy with transformation of one variable
#X_train_copy = train.iloc[:, [4]]*train.iloc[:, [4]]
X_train_copy = np.sqrt(train.iloc[:, [1]])
X_test_copy = np.sqrt(test.iloc[:, [1]])
model = skl_da.QuadraticDiscriminantAnalysis()
model.fit(X_train_copy, Y_train)
prediction = model.predict(X_test_copy)

accracy = np.mean(prediction == Y_test)
print(f"Accuracy with 1 column is :{accracy}")
```

```python
#implementation of LDA
X_train = train.iloc[:, [2,3,4,5,7,8,10,12,13]]
X_test = test.iloc[:, [2,3,4,5,7,8,10,12,13]]
print(Y_train.shape)
model1 = skl_da.LinearDiscriminantAnalysis()
model1.fit(X_train, Y_train)
prediction = model1.predict(X_test)

accracy = np.mean(prediction == Y_test)
print(f"Accuracy with 2,3,4,5,7,8,10,12,13 column is :{accracy}")
```

```python
#implementation of QDA
X_train = train.iloc[:, [2,3,4,5,7,8,10,12,13]]
X_test = test.iloc[:, [2,3,4,5,7,8,10,12,13]]
print(Y_train.shape)
model2 = skl_da.QuadraticDiscriminantAnalysis()
model2.fit(X_train, Y_train)
prediction = model2.predict(X_test)

accracy = np.mean(prediction == Y_test)
print(f"Accuracy with 2,3,4,5,7,8,10,12,13 column is :{accracy}")
```

```python
#cross validation for LDA and QDA with original data

X = data.drop(columns='Lead')
Y = data['Lead']

#print(X.describe())
#print(Y.head())
models = []
model1 = skl_da.LinearDiscriminantAnalysis()
model2 = skl_da.QuadraticDiscriminantAnalysis()
models.append(model1)
models.append(model2)

n_fold = 10

misclassification = np.zeros((n_fold, len(models)))
cv = skl_ms.KFold(n_splits=n_fold, random_state=1, shuffle=True)

for i, (train_index, val_index) in enumerate(cv.split(data)):
  X_train, X_val = X.iloc[train_index], X.iloc[val_index]
  Y_train, Y_val = Y.iloc[train_index], Y.iloc[val_index]

  for j in range(np.shape(models)[0]):
    model = models[j]
    model.fit(X_train, Y_train)
    prediction = model.predict(X_val)
    misclassification[i][j] = np.mean(prediction != Y_val)

plt.boxplot(misclassification, showmeans=True)
plt.title('Cross valication error for LDA and QDA')
plt.xticks(np.arange(2)+1, ('LDA', 'QDA'))
plt.ylabel('Validation error')
plt.show

```
